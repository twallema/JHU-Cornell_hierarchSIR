```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, lognorm
```

```python
# Data
#data = np.array([1.16E-04, 1.55E-04, 3.30E-04, 1.56E-04, 1.19E-04, 9.24E-05, 1.42E-04])
data = np.array([1.53E-04, 1.35E-04, 1.38E-04, 4.83E-05, 3.67E-05, 4.67E-05, 5.49E-05])
#data = np.array([1.60E-04, 2.51E-04, 2.61E-04, 1.11E-04, 9.64E-05, 1.28E-04, 1.43E-04])

# Take log of the data
log_data = np.log(data)

# Get parameters of the normal distribution of the logs
mu = np.mean(log_data)
sigma = np.std(log_data)

# Reconstruct the lognormal distribution using scipy
s = sigma  # shape
scale = np.exp(mu)
x = np.linspace(min(data) * 0.5, max(data) * 2, 1000)
lognorm_pdf = lognorm.pdf(x, s=s, scale=scale)

# Plot
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, density=True, alpha=0.3, label='Observed Data')
plt.plot(x, lognorm_pdf, label=f'Lognormal Fit\nμ={mu:.2f}, σ={sigma:.2f}', lw=2, linestyle='--')
plt.title('Lognormal Fit to Observed Data')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
